```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
schools_upload = "Resources/schools_complete.csv"
students_upload = "Resources/students_complete.csv"
```

```python
# Read School Data File and store into Pandas DataFrame
school_data = pd.read_csv(schools_upload)
school_data.head()
```

```python
# Read Student Data File and store into Pandas DataFrame
student_data = pd.read_csv(students_upload)
student_data.head()
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
# delete gender column -- irrelevant
del school_data_complete["gender"]
# clean up column titles
school_data_complete = school_data_complete.rename(columns={"student_name":"Student Name","school_name":"School Name","type":"Type"})
#formatting
#school_data_complete["size"] = school_data_complete ["size"].map("{:,}".format)
#school_data_complete["budget"] = school_data_complete["budget"].map("${:,.2f}".format)
school_data_complete.head()

```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# calculate total number of schools
total_schools = len(school_data_complete["School Name"].unique())

# calculate total number of students
total_students = school_data_complete["Student ID"].count()

# calculate the total budget
total_budget = school_data["budget"].sum()

# calculate the average math score
average_math = round(school_data_complete["math_score"].mean(),2)

# percentage of students with a passing math score (70 or greater)
pass_math = school_data_complete[school_data_complete["math_score"]>=70]
num_pass_math = pass_math["math_score"].count()
percent_pass_math = round((num_pass_math/total_students)*100,2)

# calculate the average reading score
average_reading = round(school_data_complete["reading_score"].mean(),2)

# percentage of students with a passing reading score (70 or greater)
pass_reading = school_data_complete[school_data_complete["reading_score"]>=70]
num_pass_reading = pass_reading["reading_score"].count()
percent_pass_reading = round((num_pass_reading/total_students)*100,2)

# calculate the percentage of students who passed math AND reading
overall_pass = school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)]
num_overall_pass = overall_pass["Student ID"].count()
percent_overall_pass = round((num_overall_pass/total_students)*100,2)
```

```python
# create a data frame with all the above info
District_Summary = pd.DataFrame({"Total Schools": [total_schools],
                                 "Total Students": [total_students],
                                 "Total Budget": [total_budget],
                                 "Average Math Score": [average_math],
                                 "% Passing Math": [percent_pass_math],
                                 "Average Reading Score": [average_reading],
                                 "% Passing Reading": [percent_pass_reading],
                                 "% Overall Passing": [percent_overall_pass]})

#formatting
District_Summary["Total Students"] = District_Summary["Total Students"].map("{:,}".format)
District_Summary["Total Budget"] = District_Summary["Total Budget"].map("${:,.2f}".format)

District_Summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# group by school name
schools = school_data_complete.groupby(["School Name"])

# school types
school_type = school_data.set_index("school_name")["type"]

# total students per school
students_per_school = schools["Student ID"].count()

# total school budget
budget_per_school = schools["budget"].first()

# per student budget
budget_per_student = budget_per_school/students_per_school

# average math score
m_avg = round(schools['math_score'].mean(),2)

# % passing math PER SCHOOL
m_pass = school_data_complete[school_data_complete["math_score"]>=70].groupby(["School Name"]).size()
percent_m_pass = round((m_pass/students_per_school)*100,2)

# average reading score PER SCHOOL
r_avg = round(schools['reading_score'].mean(),2)

# % passing reading PER SCHOOL
r_pass = school_data_complete[school_data_complete["reading_score"]>=70].groupby(["School Name"]).size()
percent_r_pass = round((r_pass/students_per_school)*100,2)

# % overall passing PER SCHOOL
o_pass = school_data_complete[school_data_complete["math_score"]>=70].groupby(["School Name"]).size() & school_data_complete[school_data_complete["reading_score"]>=70].groupby(["School Name"]).size()
percent_o_pass = round((o_pass/students_per_school)*100,2)


# create new dataframe containing above info
School_Summary = pd.DataFrame({"School Type":school_type,
                               "Students Per School":students_per_school,
                               "School Budget":budget_per_school,
                               "Per Student Budget":budget_per_student,
                               "Average Math Score":m_avg,
                               "% Passing Math":percent_m_pass,
                               "Average Reading Score":r_avg,
                               "% Passing Reading":percent_r_pass,
                               "% Overall Passing":percent_o_pass})

#formatting
#School_Summary["Students Per School"] = School_Summary["Students Per School"].map("{:,}".format)
#School_Summary["School Budget"] = School_Summary["School Budget"].map("${:,.2f}".format)
School_Summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# extract relevant fields from School Summary for next 3 tables
scores = School_Summary[["Average Math Score",
                        "% Passing Math",
                        "Average Reading Score",
                        "% Passing Reading",
                        "% Overall Passing"]]
```

```python
# create bins for per student budget
bins = [0, 585, 630, 645, 680]

# create names for the bins
group_names = ["<$585", "$585-630", "$631-645", "$646-680"]

# add new column w/ bins
school_data_complete["Budget Range (Per Student)"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], bins, labels=group_names, include_lowest=True)

# group by "Budget Range (Per Student)"
budget_bins = school_data_complete.groupby("Budget Range (Per Student)")

# calculations
m_avg = round(budget_bins["math_score"].mean(),2)

r_avg = round(budget_bins["reading_score"].mean(),2)

m_pass = school_data_complete[school_data_complete["math_score"]>=70].groupby("Budget Range (Per Student)")["Student ID"].count()
percent_m_pass = round((m_pass/budget_bins["Student ID"].count())*100,2)

r_pass = school_data_complete[school_data_complete["reading_score"]>=70].groupby("Budget Range (Per Student)")["Student ID"].count()
percent_r_pass = round((r_pass/budget_bins["Student ID"].count())*100,2)
                                                                                                            
o_pass = (percent_m_pass + percent_r_pass) / 2


# budget dataframe
budget_range = pd.DataFrame({"Average Math Score":m_avg,
                               "% Passing Math":percent_m_pass,
                               "Average Reading Score":r_avg,
                               "% Passing Reading":percent_r_pass,
                               "% Overall Passing":o_pass})
    
budget_range
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# create bins for school size
bins = [0, 1499, 2499, 5000]

# create group names for the bins
group_names = ["Small (<1500)", "Medium (1500-2499)", "Large (2500-5000)"]

# add new column w/ bins
school_data_complete["School Size"] = pd.cut(school_data_complete["size"], bins, labels=group_names, include_lowest=True)


# group by School Size
size_bins = school_data_complete.groupby("School Size")

# calculations
m_avg = round(size_bins["math_score"].mean(),2)

r_avg = round(size_bins["reading_score"].mean(),2)

m_pass = school_data_complete[school_data_complete["math_score"]>=70].groupby("School Size")["Student ID"].count()
percent_m_pass = round((m_pass/size_bins["Student ID"].count())*100,2)

r_pass = school_data_complete[school_data_complete["reading_score"]>=70].groupby("School Size")["Student ID"].count()
percent_r_pass = round((r_pass/size_bins["Student ID"].count())*100,2)

o_pass = (percent_m_pass + percent_r_pass) / 2

# size dataframe
size_range = pd.DataFrame({"Average Math Score":m_avg,
                               "% Passing Math":percent_m_pass,
                               "Average Reading Score":r_avg,
                               "% Passing Reading":percent_r_pass,
                               "% Overall Passing":o_pass})
    
size_range
```
